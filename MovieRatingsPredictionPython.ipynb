```python
budget.head()
```

```python
#lets visualize the budget of movies just for fun
import matplotlib.pyplot as plt
%matplotlib inline
budget[["movie_name","production_budget","worldwide_gross"]]
# format isn't very useful so we convert it to 
```

```python
#the format isn't useful for plotting even if it looks good
budget_numeric = budget[['domestic_gross', 'production_budget', 'worldwide_gross']].replace('[\$,]', '', regex=True).astype(float)
```

```python
#lets visualize budgets against gross

ax1 = plt.subplot(311)
plt.plot(budget_numeric.production_budget)
plt.setp(ax1.get_xticklabels(), fontsize=12)
ax2 = plt.subplot(312,sharex=ax1)
plt.plot(budget_numeric.domestic_gross)
plt.setp(ax2.get_xticklabels(), fontsize=12)
ax3 = plt.subplot(313,sharex=ax1)
plt.plot(budget_numeric.worldwide_gross)
plt.setp(ax3.get_xticklabels(), fontsize=12)
plt.show()
```

```python
#from the above looks like budgets are about consistant while there is a lot of varience in gross
#lets confirm using the describe function
print(budget_numeric.domestic_gross.describe())
print()
print(budget_numeric.worldwide_gross.describe())
print()
print(budget_numeric.production_budget.describe())
#given the above, the theory is confirmed.
#additionally we learn that there is an anomaly in both domestic_gross and worldwide_gross where the minimum value is 0
#we may want to fix this later but my theory is that this is just incomplete data
```
